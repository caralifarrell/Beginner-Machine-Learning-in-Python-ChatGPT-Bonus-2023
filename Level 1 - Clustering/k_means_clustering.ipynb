## Training the K-Means model on the dataset

```python
# Create K-Means model
kmeans = KMeans(n_clusters = 5, init="k-means++", random_state = 42)

# Train the model and returns the 5 clusters themselves in the form of a vector.
# Each of the observations of the data set (x) will be shown the index of the cluster it 
# belongs to.
y_pred = kmeans.fit_predict(X)
y_pred
```
